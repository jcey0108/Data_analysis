```python
#折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams, font_manager

# 一次性设置全局中文字体（宋体）
font_path = '/System/Library/Fonts/Supplemental/Songti.ttc'
rcParams['font.family'] = font_manager.FontProperties(fname=font_path).get_name()

# 可选：解决负号乱码（常见于坐标轴）
rcParams['axes.unicode_minus'] = False

#设置画布大小
plt.figure(figsize=(10, 5))

#设置要绘图的数据
month = ['1月','2月','3月', '4月']
sales = [100, 200, 300, 20]

#画折线图
plt.plot(month, sales, label='产品A', marker='o') #产品A是图例
#添加标题
plt.title('2025年销售趋势', color='red')
#添加坐标轴标签
plt.xlabel('月份', fontsize=10)
plt.ylabel('销售额（万元）',fontsize=10)

#添加图例
plt.legend(loc='upper left')
#添加网格线
plt.grid(True, alpha=0.1, color='green', linestyle='--')
#设置刻度字体大小
plt.xticks(rotation=40, fontsize = 12) #刻度旋转
plt.xticks(rotation=0, fontsize = 12)
#设置y轴范围
plt.ylim(0, 500)
#在每个数据点上显示数值
for x, y in zip(month, sales):
    print(x, y)
    plt.text(x, y+10, str(y), ha = 'center', va = 'bottom')

#显示图表
plt.show()
```

```python
#柱状图
import matplotlib.pyplot as plt
from matplotlib import rcParams, font_manager

# 一次性设置全局中文字体（宋体）
font_path = '/System/Library/Fonts/Supplemental/Songti.ttc'
rcParams['font.family'] = font_manager.FontProperties(fname=font_path).get_name()

# 可选：解决负号乱码（常见于坐标轴）
rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 5))
subjects = ['语文','数学','英语','科学']
scores = [102,98, 66, 100]
#柱状图
plt.bar(subjects, scores,label='小红')
plt.title("成绩分布")
plt.xlabel('科目', fontsize = 10)
plt.ylabel('分数', fontsize = 10)
#添加网格线
plt.grid()

#添加图例
plt.legend(loc='upper right')
plt.show()
```

```python
#绘制饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams, font_manager

# 一次性设置全局中文字体（宋体）
font_path = '/System/Library/Fonts/Supplemental/Songti.ttc'
rcParams['font.family'] = font_manager.FontProperties(fname=font_path).get_name()

# 可选：解决负号乱码（常见于坐标轴）
rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 5))
things = ['学习','娱乐','睡觉','其他']
times = [4, 3, 8, 9]
colors=['#66b3ff', '#99ff99', "#ffcc99", '#ff9999', '#ff4499']
plt.pie(times,labels = things, autopct='%.1f',
        startangle=90, colors=colors) #调整初始画图的角度
plt.title('一天时间分布')
plt.show()
```

```python
#环形图
import matplotlib.pyplot as plt
from matplotlib import rcParams, font_manager

# 一次性设置全局中文字体（宋体）
font_path = '/System/Library/Fonts/Supplemental/Songti.ttc'
rcParams['font.family'] = font_manager.FontProperties(fname=font_path).get_name()

# 可选：解决负号乱码（常见于坐标轴）
rcParams['axes.unicode_minus'] = False

plt.figure(figsize=(10, 5))

things = ['学习','娱乐','睡觉','其他']
times = [4, 3, 8, 9]
colors=['#66b3ff', '#99ff99', "#ffcc99", '#ff9999', '#ff4499']
plt.pie(times,labels = things, autopct='%.1f',
        startangle=90, colors=colors, wedgeprops={'width':0.5}) #调整初始画图的角度
plt.title('一天时间分布')
plt.show()
```

```python
#分析案例，温度分析
#1.导入库
import pandas as pd
import matplotlib.pyplot as plt
# 一次性设置全局中文字体（宋体）
font_path = '/System/Library/Fonts/Supplemental/Songti.ttc'
rcParams['font.family'] = font_manager.FontProperties(fname=font_path).get_name()

# 可选：解决负号乱码（常见于坐标轴）
rcParams['axes.unicode_minus'] = False

#2.导入数据
df = pd.read_csv('data/weather.csv')


#绘制一个气温的趋势变化图
df['date'] = pd.to_datetime(df['date'])
df = df[df['date'].dt.year==2015]

plt.figure(figsize=(10, 10))
plt.plot(df['date'], df['temp_max'], label = '最高气温')
plt.plot(df['date'], df['temp_min'], label = '最低气温')

plt.title('2015年')
plt.xlabel('日期')
plt.ylabel('气温')
plt.legend()








```

```python
#绘制降水量的直方图
plt.hist(df['precipitation'], bins=5)
```
